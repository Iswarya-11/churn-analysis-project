converting the senior citizen o and 1 into=>Yes,No

```python
def data(value):
  if value==1:
    return "Yes"
  else:
    return "No"
df['SeniorCitizen']=df['SeniorCitizen'].apply(data)
```

```python
df['Churn']=df['Churn'].str.strip().str.capitalize()
```

```python
df['Churn'].unique()
df['Churn'].value_counts()
```

```python
df['Churn']=df['Churn'].astype(str)
df['Churn']=df['Churn'].str.strip()
df['Churn']=df['Churn'].str.title()
```

```python
df['Churn'].unique()
df['Churn'].value_counts()
```

```python
ax=sns.countplot(x = 'Churn',data = df)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
plt.figure(figsize=(3,4))    #Figure size()
gb=df.groupby('Churn').agg({'Churn':'count'})
plt.pie(gb['Churn'],labels=gb.index,autopct='%1.2f%%')
plt.title("PERCENTAGE OF CHURNED CUSTOMER",fontsize=10)   #font size()
plt.show()
gb
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x="gender",data=df,hue='Churn')
plt.title("Churn By Genders")
plt.show()
```

```python
ax=sns.countplot(x='SeniorCitizen',data=df)
ax.bar_label(ax.containers[0])
plt.title("Count Of Customer By Senior Citizen")
plt.show()
```

```python
# Step 1: Create a crosstab of SeniorCitizen vs Churn
crosstab = pd.crosstab(df['SeniorCitizen'], df['Churn'], normalize='index') * 100

# Step 2: Plot the stacked bar chart
ax = crosstab.plot(kind='bar', stacked=True, colormap='Set2', figsize=(8, 6))

# Step 3: Add percentage labels
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    if height > 0:
        ax.text(x + width/2, y + height/2,
                f'{height:.1f}%', ha='center', va='center', fontsize=10)

# Step 4: Formatting
plt.title("Churn by Senior Citizen ", fontsize=14)
plt.ylabel("Percentage")
plt.xlabel("Senior Citizen")
plt.xticks([0, 1], labels=["Not Senior", "Senior"], rotation=0)
plt.legend(title="Churn", loc='upper right')
plt.ylim(0, 100)
plt.tight_layout()

# Step 5: Show the plot
plt.show()
```
